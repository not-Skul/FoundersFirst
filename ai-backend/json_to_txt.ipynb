```python
for count, item in enumerate(data):
    print(f"{count+1} is {item['scheme_name']}")
```

```python
with open('central_schemes.txt','w',encoding='utf-8') as f:
    for scheme in data:
        f.write("scheme name - " + scheme['scheme_name'] + '\n')
        f.write("ministry - " + scheme['ministry'] + '\n')
        f.write("department - " + scheme['department'] + '\n')
        f.write("key sectors - " + scheme['key_sectors'] + '\n')
        f.write("brief - " + scheme['brief'] + '\n')
        f.write("eligibility - " + scheme['eligibility'] + '\n')
        f.write("benefits - " + scheme['benefits'] + '\n')
        f.write("benefit tags - " + scheme['benefit_tags'] + '\n')
        f.write("tenure - " + scheme['tenure'] + '\n')
        f.write("application link - " + scheme['application_link'] + '\n\n')
```
